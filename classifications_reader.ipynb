```python
import pandas as pd
import json
import os
```

```python
os.chdir('./classified_json_files')
json_files = os.listdir('.')
```

```python
df = pd.DataFrame()
```

```python
for file in json_files:
    df = pd.concat([df, pd.read_json(file)])
```

```python
df = df.explode('type')
```

```python
df.to_csv('classified_exercises.csv', index = False)
```
