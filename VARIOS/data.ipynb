```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
add_df = pd.read_csv('classified_json_files/classified_exercises.csv')
```

Function to create the exercise string from `exam` column values.

```python
def form_exercise_string(row):
    # exam = [row['exam'].upper() ]
    exercise = [i.upper() for i in row['exam'].split(', ')]
    return '. '.join(exercise)
```

```python
add_df['exercise'] = add_df.apply(form_exercise_string, axis = 1)
```

Extract exam from `exercise` string

```python
add_df['exam'] = add_df['exam'].apply(lambda x : x.split(', ')[0])
```

Rename `type` to `exercise_type`

```python
add_df.rename(columns = {'type' : 'exercise_type'}, inplace=True)
```

Add `statement` column

```python
add_df['statement'] = None
```

Reaarrange the columns to match the columns in the base dataframe

```python
# subject,year,topic,exam,exercise_number,statement,exercise_type
add_df = add_df[['subject', 'year', 'topic', 'exam', 'exercise', 'exercise_type', 'statement']]
```

Check if column structure matches the base dataframe

```python
add_df.head(2)
```

```python
df.head(2)
```

Merge dataframes

```python
df = pd.concat([df, add_df])
```

Remove duplicates

```python
df = df.drop_duplicates()
```

Clean `Química` subject name

```python
df.loc[df.subject.isin(['QUÍMICA', 'QUIMICA', 'Química']), 'subject'] = 'Química'
```

Export the merged dataframe to `data.csv`

```python
df.to_csv('data.csv', index = False)
```
