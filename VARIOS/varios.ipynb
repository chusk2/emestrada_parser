```python
from pathlib import Path
import json
```

```python
folders = {
    'Física' : "/home/daniel/datos2/SELECTIVIDAD/Física/Fisica - Temas",
    'Química' : "/home/daniel/datos2/SELECTIVIDAD/Química/Química - Temas",
    'Mates II' : "/home/daniel/datos2/SELECTIVIDAD/Mates II/Mates II - Temas",
    "Mates CCSS" : "/home/daniel/datos2/SELECTIVIDAD/Mates CCSS/Mates CCSS - Temas"
}
```

```python
SUBJECTS = ['Física', 'Química', 'Mates II', "Mates CCSS"]
```

```python
topics = {}
```

```python
for subject, folder in folders.items():
    path = Path(folder)
    directories = path.iterdir()
    topics[subject] = [i.name for i in directories if i.is_dir()]
```

```python
topics
```

```python
with open('subject_topics.json', 'w') as file:
    json.dump(topics, file)
```

```python
with open('./data/subject_topics.json', 'r') as file:
    exercise_types_dict = json.load(file)
```

```python
exercise_types = {subject : {t : [] for t in topics}  for subject, topics in exercise_types_dict.items()}
```
