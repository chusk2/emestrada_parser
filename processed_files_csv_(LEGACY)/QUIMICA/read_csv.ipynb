```python
exercises_files = [i for i in pathlib.Path(".").iterdir()
                   if i.name.startswith("exercises_")]
```

```python
for file in exercises_files:
    df = pd.read_csv(file)
    print(f"{list(df.columns)} --> headers {file.name.split("exercises_")[1]}")
```

```python
data = pd.read_csv(exercises_files[0])
for file in exercises_files[1:]:
    data = pd.concat([data, pd.read_csv(file)], axis=0)
```
