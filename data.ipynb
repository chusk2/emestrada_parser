```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
add_df = pd.read_csv('classified_json_files/classified_exercises.csv')
```

```python
df.head()
```

```python
add_df
```

```python
def form_exercise_string(row):
    # exam = [row['exam'].upper() ]
    exercise = [i.upper() for i in row['exam'].split(', ')]
    return '. '.join(exercise)
```

```python
add_df['exercise'] = add_df.apply(form_exercise_string, axis = 1)
```

```python
add_df['exam'] = add_df['exam'].apply(lambda x : x.split(', ')[0])
```

```python
add_df.rename(columns = {'type' : 'exercise_type'}, inplace=True)
```

```python
add_df['statement'] = None
```

```python
# subject,year,topic,exam,exercise_number,statement,exercise_type
add_df = add_df[['subject', 'year', 'topic', 'exam', 'exercise', 'exercise_type', 'statement']]
```

```python
add_df.head()
```

```python
df.head()
```

```python
df = pd.concat([df, add_df])
```

```python
df.to_csv('data.csv', index = False)
```
