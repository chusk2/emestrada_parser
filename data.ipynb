```python
import pandas as pd
```

```python
df = pd.read_csv('data.csv')
```

```python
df.head()
```

```python
def form_exercise_string(row):
    exam = [row['exam'].upper() ]
    exercise = [i.upper() for i in row['exercise_number'].split(', ')]
    return '. '.join(exam + exercise)
```

```python
df['exercise'] = df.apply(form_exercise_string, axis = 1)
```

```python
df.sample(10)
```

```python
# subject,year,topic,exam,exercise_number,statement,exercise_type
df = df[['subject', 'year', 'topic', 'exam', 'exercise', 'exercise_type', 'statement']]
```

```python
df.to_csv('data.csv', index = False)
```
