```python
from PyPDF2 import PdfReader
import pandas as pd
```

```python
def get_exams(file):
    pdf_file = PdfReader('equilibrio.pdf')
    first_page = pdf_file.pages[0]
    text = first_page.extract_text().split('\n')

    exams = []

    for line in text:
        line = line.strip()
        if line.startswith('20'):
            year = int(line)
        elif line in ['QUÍMICA', 'FISICA', 'MATEMÁTICAS']:
            subject = line
        elif line.startswith('TEMA'):
            topic = line.split(': ')[1]
        elif line[0] == '•':
            exams.append(line)
        else:
            continue

    return year, subject, topic, exams
```
